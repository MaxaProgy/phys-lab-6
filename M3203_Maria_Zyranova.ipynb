## Затухающие колебания в LCR контуре.


```python
import plotly.graph_objects as go
import numpy as np
```

```python
# @title Параметры
# @markdown Индуктивность, L (Гн):
L = 10 # @param {type:"raw"}
# @markdown Емкость, C (Ф):
C = 0.01 # @param {type:"raw"}
# @markdown Сопротивление, R (Ом):
R = 5 # @param {type:"raw"}
# @markdown Время, t (c):
t = 10 # @param {type:"integer"}
# @markdown Начальная сила заряда, q0 (Кл):
q_0 = 5 # @param {type:"raw"}
# @markdown Начальная сила тока, I0 (А):
I_0 = 10 # @param {type:"raw"}
# @markdown Начальное напряжение, U0 (В):
U_0 = 15 # @param {type:"raw"}
```

```python
if L < 0:
    raise Exception("Введите неотрицательную индуктивность")
if t <= 0:
    raise Exception("Введите положительное время")
if R < 0:
    raise Exception("Введите неотрицательное сопротивление")
if C < 0:
    raise Exception("Введите неотрицательную емкость")
if q_0 < 0:
    raise Exception("Введите неотрицательную силу заряда")
if I_0 < 0:
    raise Exception("Введите неотрицательную начальную силу тока")
if U_0 < 0:
    raise Exception("Введите неотрицательное начальное напряжение")
```

```python
step = t / 1000
b = R / (2 * L)
w = np.sqrt(1 / (L * C) - (R ** 2) / (4 * L ** 2))

if np.isnan(w):
    raise Exception("Частота не может быть отрицательной")

x = np.linspace(0, t, 1000)

q = q_0 * np.exp(-b * x) * np.cos(w * x)
I = I_0 * np.exp(-b * x) * np.sin(w * x)
U = U_0 * np.exp(-b * x) * np.cos(w * x)
```

## Сила заряда

```python
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=x, y=q, mode='lines+markers', marker_color='blue'))
fig1.update_layout(width=800, height=320, xaxis_title='Время, с', yaxis_title='Сила заряда, Кл')
fig1.show()
```

## Сила тока

```python
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=x, y=I, mode='lines+markers', marker_color='red'))
fig2.update_layout(width=800, height=300, xaxis_title='Время, с', yaxis_title='Сила тока, А')
fig2.show()
```

## Напряжение

```python
fig3 = go.Figure()
fig3.add_trace(go.Scatter(x=x, y=U, mode='lines+markers', marker_color='green'))
fig3.update_layout(width=800, height=300, xaxis_title='Время, с', yaxis_title='Напряжение, В')
fig3.show()
```
